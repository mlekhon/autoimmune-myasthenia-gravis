# Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
df_original= pd.read_csv('imputed_dataset.csv')
df_original.head()
```

# Correlation Heatmap Matrix

```python
# --- 1. Rename columns in the original dataframe ---

rename_dict = {
    'age': 'age',
    'age_erst': 'age_at_onset',
    'age_diag': 'age_at_diagnosis',
    'time_erst_to_diag': 'time_from_onset_to_diagnosis',
    'dauer': 'disease_duration',
    'chronicfatigue_neu': 'chronic_fatigue_score',
    'scoreqol_neu': 'qol_score',
    'scoreadl_neu': 'adl_score',
    'mgfaklassifikation_schlimmste_historisch': 'mgfa_worst_history',
    'mgfaklassifikation_schlimmste_historisch_1bis5': 'mgfa_worst_history_1to5',
    'aktueller_mgfa_score': 'mgfa_current_score',
    'MGFA_umkodiert_betonung': 'mgfa_recoded_emphasis',
    'mgfa_aktuell_1bis5': 'mgfa_current_1to5',
    'japan_classification': 'japan_classification',
    'ee_auswahlbildungsabschluss_rb': 'education_level',
    'schwerbehinderung': 'severe_disability'
}

df_original = df_original.rename(columns=rename_dict)

# --- 2. Define continuous and ordinal variables using new names ---

continuous_vars = [
    'age',
    'age_at_onset',
    'age_at_diagnosis',
    'time_from_onset_to_diagnosis',
    'disease_duration',
    'chronic_fatigue_score',
    'qol_score',
    'adl_score'
]

ordinal_vars = [
    'mgfa_worst_history',
    'mgfa_worst_history_1to5',
    'mgfa_current_score',
    'mgfa_recoded_emphasis',
    'mgfa_current_1to5',
    'japan_classification',
    'education_level',
    'severe_disability'
]

# --- 3. Compute Pearson correlation for continuous variables ---

pearson_corr = df_original[continuous_vars].corr(method='pearson')

# --- 4. Compute Spearman correlation for all variables ---

spearman_corr = df_original[continuous_vars + ordinal_vars].corr(method='spearman')

# --- 5. Build combined correlation matrix (Pearson for cont-cont, Spearman otherwise) ---

combined_corr = spearman_corr.copy()

for var1 in continuous_vars:
    for var2 in continuous_vars:
        if var1 != var2:
            combined_corr.loc[var1, var2] = pearson_corr.loc[var1, var2]

# --- 6. Plot heatmap ---

mask = np.triu(np.ones_like(combined_corr, dtype=bool))

plt.figure(figsize=(12, 10))
sns.heatmap(
    combined_corr,
    mask=mask,
    cmap="Blues",
    vmin=-1,
    vmax=1,
    annot=True,
    fmt=".2f",
    cbar_kws={"shrink": .8},
    linewidths=0.5,
    linecolor='darkgrey'
)

plt.title("Correlation Heatmap (Pearson for Continuous-Continuous, Spearman Otherwise)")
plt.tight_layout()
plt.show()
```

```python
# Define continuous and ordinal variables
continuous_vars = ['age',
    'age_at_onset',
    'age_at_diagnosis',
    'time_from_onset_to_diagnosis',
    'disease_duration',
    'chronic_fatigue_score',
    'qol_score',
    'adl_score']

ordinal_vars = ['mgfa_worst_history',
    'mgfa_worst_history_1to5',
    'mgfa_current_score',
    'mgfa_recoded_emphasis',
    'mgfa_current_1to5',
    'japan_classification',
    'education_level',
    'severe_disability']

# Compute Pearson correlation for continuous variables
pearson_corr = df_original[continuous_vars].corr(method='pearson')

# Compute Spearman correlation for all variables
spearman_corr = df_original[continuous_vars + ordinal_vars].corr(method='spearman')

# Initialize the final correlation matrix with Spearman values
combined_corr = spearman_corr.copy()

# Replace continuous-continuous pairs with Pearson correlation values
for var1 in continuous_vars:
    for var2 in continuous_vars:
        if var1 != var2:
            combined_corr.loc[var1, var2] = pearson_corr.loc[var1, var2]

# Filter for correlations with absolute values greater than 0.4
filtered_corr = combined_corr[(combined_corr > 0.4) | (combined_corr < -0.4) & (combined_corr != 1.0)]

# Create a mask for the upper triangle to avoid redundancy in the heatmap
mask = np.triu(np.ones_like(filtered_corr, dtype=bool))

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(filtered_corr, mask=mask, cmap="Blues", vmin=-1, vmax=1, annot=True, fmt=".2f", cbar_kws={"shrink": .8}, linewidths=0.5, linecolor='darkgrey')

plt.title("Correlation Heatmap (Pearson for Continuous-Continuous, Spearman Otherwise, |Corr| > 0.4)")
plt.tight_layout()
plt.show()
```
